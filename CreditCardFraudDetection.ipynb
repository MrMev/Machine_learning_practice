# Credit Card Fraud Detection


### Importing libraries


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

#### loding the dataset to Pandas DataFrame

```python
credit_card_data = pd.read_csv("creditcard.csv")
```

```python
## first % rows of dataset
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
# dataset information
credit_card_data.info()
```

```python
# some description about dataset
credit_card_data.describe()
```

```python
# distibution of legit transaction & fraudulent transactions
credit_card_data['Class'].value_counts()
```

#### This Dataset is highly unblanced

- 0 --> Normal Transaction

- 1 --> fraudulent transaction

```python
#sperating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# Compare the values for both transaction
credit_card_data.groupby('Class').mean()
```

#### Under-Sampling
 Build a sample dataset contaning similar disrributation of normal transction and fraud transaction
 - Number of Fraudulent Transaction -- 492

```python
# making a sample dataset from legit transaction 
legit_sample = legit.sample(n=492)
```

#### Concatenating two DataFrame
To make even dataset that will result in better analysis

```python
new_dataset = pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

As we can get approximately same mean for the actual and sample dataset that yeild the nature of the dataset is similar So if we serve it to machine learning model it will detect the fraud and legit transaction


#### Splitting the data into Features and Traget(0&1)

```python
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

#### Split the data into training & testing data

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

### Model Traning 
- ###### Logistic Regression

```python
model = LogisticRegression()
```

```python
#training the Logistic Regression Model with Training Data
model.fit(X_train,Y_train)
```

### Model Evaluation 
- ###### Accuracy Score

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy score on Training Data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```
